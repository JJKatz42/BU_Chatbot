```python
import openai
import pandas as pd
import os
import html2text
import time

# Directory containing HTML files
DIRECTORY = "/workspaces/BU_Chatbot/Questrom_Course_Info"

openai.api_key = "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"

# Get list of all HTML files
html_files = [file for file in os.listdir(DIRECTORY) if file.endswith(".html")]

print(html_files)

# Dictionary to store each file's content
html_contents = {}

for html_file in html_files:
    with open(f'{DIRECTORY}/{html_file}', 'r') as f:
        html_content = f.read()
        h1 = html2text.HTML2Text()
        h1.ignore_links = True
        h2 = h1.handle(html_content)
        html_contents[html_file] = h2

print (html_contents)

api_request_counter = 0

# Get embeddings for each html file
f_titles = []
embeddings = []

for html_file, content in html_contents.items():
    embedding = openai.Embedding.create(input=content, engine='text-embedding-ada-002')['data'][0]['embedding']
    embeddings.append(embedding)
    f_titles.append(html_file)
    print(len(embedding))

    api_request_counter += 1
    print(api_request_counter)

    # Adding delay here
    time.sleep(1)

print(embeddings)

# Creating the DataFrame
df = pd.DataFrame({'title': f_titles, 'text': list(html_contents.values())[:len(embeddings)], 'embeddings': embeddings})

df.to_csv('/workspaces/BU_Chatbot/open_ai_test/embeddings.csv')

df.head()
```
